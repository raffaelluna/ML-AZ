```python
import pandas as pd
import numpy as np
```

```python
path = '~/ML-AZ/plano_saude.csv'
df = pd.read_csv(path)
```

```python
X = df.iloc[:,0].values
y = df.iloc[:,1].values
```

```python
corr = np.corrcoef(X,y)
print(corr)
```

```python
X = X.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
print(reg.intercept_)
print(reg.coef_)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X,y)
plt.plot(X, reg.predict(X), color='red')
plt.title('Linear Regression')
plt.xlabel('Idade')
plt.ylabel('Custo')
```

```python
prev = reg.intercept_ + reg.coef_ * 40
prev
```

```python
score = reg.score(X, y)
score
```

```python
from yellowbrick.regressor import ResidualsPlot
```

```python
viz = ResidualsPlot(reg)
viz.fit(X,y)
viz.poof()
```
