```python
import pandas as pd
import numpy as np
```

```python
path = '~/ML-AZ/plano_saude.csv'
df = pd.read_csv(path)
```

```python
X = df.iloc[:,0].values
y = df.iloc[:,1].values
```

```python
corr = np.corrcoef(X,y)
print(corr)
```

```python
X = X.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
clf = LinearRegression()
clf.fit(X,y)
```

```python
print(clf.intercept_)
print(clf.coef_)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X,y)
plt.plot(X, clf.predict(X), color='red')
plt.title('Linear Regression')
plt.xlabel('Idade')
plt.ylabel('Custo')
```

```python
prev = clf.intercept_ + clf.coef_ * 40
prev
```

```python
score = clf.score(X, y)
score
```

```python
from yellowbrick.regressor import ResidualsPlot
```

```python
viz = ResidualsPlot(clf)
viz.fit(X,y)
viz.poof()
```
